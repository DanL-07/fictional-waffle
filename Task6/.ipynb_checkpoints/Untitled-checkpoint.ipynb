```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import scipy.stats as stats
import matplotlib as slop

hdulist = fits.open('m101.fits')
hdulist.info()

hdr = hdulist[0].header
keys=[hdr.keys()]
print(keys[:5])

dat = hdulist[0].data
print("dtype", dat.dtype)
print("dShape:",dat.shape)

earth=dat.flatten()
min,max,mean,std,median,mode= np.min(earth),np.max(earth),np.mean(earth),np.std(earth),np.median(earth),stats.mode(earth,keepdims=True).mode[0]
print(f' min={min}, max={max}, mean={mean:.2e}, std={std:.2e}, meadian={median},mode={mode}')

ax=plt.hist(earth, 67)
slop.ax.Axes.axvline(x=mean)
```
