```python
import numpy as np
func= input("func=: ")
a=float(input("lowerbound:"))
b=float(input("upperbound:"))

x = np.random.uniform(a, b, 100000)
f = eval(func)

integral = (b - a) * np.mean(f)

print("Montecarlo integral=", integral)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])

z = x + 1j*y
r = np.exp(1j*np.deg2rad(30))
rot = r*z

plt.plot(x, y)
plt.plot(np.real(rot), np.imag(rot))
plt.gca().set_aspect(1)#Get current axes and make them equal
plt.show() #Apparently it isn't needed for JupyterHub but it might in python, so good practice to remember
```

```python
def funcfilter(func, seq):
    seq = np.asarray(seq)
    seq = seq[seq > 0]
    return func(seq)
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)
result = funcfilter(np.log10, 100)
print("Result log10:", result)
```

```python
import numpy as np
from scipy.integrate import quad
from astropy.cosmology import Planck18

H0, Om0, Ode0 = Planck18.H0.value, Planck18.Om0, Planck18.Ode0

H = lambda z: H0*np.sqrt(Om0*(1+z)**3 + Ode0)
f = lambda z: 1/((1+z)*H(z))

I,err = quad(f,0,2) #Quad returns two valuews, we want to get both value and error
tL = I*3.085677581e19/(3600*24*365.25*1e9)

print(tL,"Gyr")
```
