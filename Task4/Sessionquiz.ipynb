```python
#4.3
import numpy as np
rng=np.random.default_rng()
w=input("input R/P/S for rock/paper/scissors",)
rand=int(rng.integers(low=0, high=3))

if w=="R":#Rock
    w=0
    if w==rand:
        print("RdR DRAW")
    elif (w+1)==rand:
        print("PbR LOSS")
    else:
        print("RbS WIN")
        
elif w=="P": #PAPER
    w=1
    if w==rand:
        print("PdP DRAW")
    elif (w+1)==rand:
        print("SbP LOSS")
    else:
        print("PbR WIN")
        
elif w=="S": #SCISSORS
    w=2
    if w==rand:
        print("SdS DRAW")
    elif (w-2)==rand:
        print("RbS LOSS")
    else:
        print("SbP WIN") #MAYBE USE np.random.choice()
```

```python
import numpy as np
def funcfilter(func, seq):
    seq = np.asarray(seq)
    seq = seq[seq > 0]
    return func(seq)
sq = np.array([-100, -10, 0, 10, 100])
result = funcfilter(np.log10, sq)
print("Result log10:", result)
result = funcfilter(np.log10, 100)
print("Result log10:", result)
```

```python
import numpy as np
from scipy.integrate import quad
from astropy.cosmology import Planck18

H0, Om0, Ode0 = Planck18.H0.value, Planck18.Om0, Planck18.Ode0

H = lambda z: H0*np.sqrt(Om0*(1+z)**3 + Ode0)
f = lambda z: 1/((1+z)*H(z))

I,err = quad(f,0,2)
tL = I*3.085677581e19/(3600*24*365.25*1e9)

print(tL,"Gyr")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Reproducibility
np.random.seed(42)

# Parameters
n = 100000
mu_x, mu_y = 0, 2
sigma = 1
bins = 30

# Generate data
x = np.random.normal(mu_x, sigma, n)
y = np.random.normal(mu_y, sigma, n)

# Create figure
fig, ax = plt.subplots(figsize=(10,6))
 #I don't know the name of the background, plot is the same tho


# Histogram
plt.hist(x, bins=bins, density=True, alpha=0.6, color='royalblue',
         edgecolor='black', label='X')
plt.hist(y, bins=bins, density=True, alpha=0.6, color='red',
         edgecolor='black', label='Y')


xmin = min(x.min(), y.min())
xmax = max(x.max(), y.max())
bin_edges = np.linspace(xmin, xmax, bins + 1)
xs = np.linspace(xmin, xmax, 500)

# PDFs
pdf_x = norm.pdf(xs, mu_x, sigma)
pdf_y = norm.pdf(xs, mu_y, sigma)

# Plot PDFs
plt.plot(xs, pdf_x, 'k:', linewidth=4, label='X(PDF)')
plt.plot(xs, pdf_y, 'k--', linewidth=4, label='Y(PDF)')

# Labels and title
plt.title('1D Histogram + PDF', fontsize=16)
plt.xlabel('Values', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)

# Annotations

ax.annotate(r'$\mathcal{N}(0,1)$',
            xy=(0, 0.15),        # arrow points here
            xytext=(-3.5, 0.145), # same y → horizontal arrow
            arrowprops=dict(facecolor='blue', width=2),
            color='blue', fontsize=14)

ax.annotate(r'$\mathcal{N}(2,1)$',
            xy=(2, 0.15),
            xytext=(4.5, 0.145),
            arrowprops=dict(facecolor='red', width=2),
            color='red', fontsize=14)
# Legend
plt.legend()

# Grid formatting
ax.set_yticks(np.arange(0, 0.45, 0.05))
ax.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace with your X and Y)
np.random.seed(0)
x = np.random.normal(0, 1, 100000)
y = np.random.normal(2, 1, 100000)

# Create 2D histogram
bins = 30
H, xedges, yedges = np.histogram2d(x, y, bins=bins)

# Transpose for plotting
H = H.T

# Meshgrid for contour plotting
Xc, Yc = np.meshgrid((xedges[:-1] + xedges[1:]) / 2,(yedges[:-1] + yedges[1:]) / 2)

# Sigma levels (approximate percentiles for 2D Gaussian)
# Convert histogram to probability
pdf = H / np.sum(H)

# Sort values descending
pdf_sorted = np.sort(pdf.ravel())[::-1]
cdf = np.cumsum(pdf_sorted)

# Find density thresholds corresponding to probability mass
levels = []
for p in [0.68, 0.95, 0.997]:
    levels.append(pdf_sorted[np.searchsorted(cdf, p)])

levels = np.sort(levels)

# Plot
plt.figure(figsize=(8,6))
# 2D histogram
plt.imshow(
    H,
    origin='lower',
    cmap='Blues',
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    aspect='auto'
)
plt.colorbar(label="Counts")
# Contours
contours = plt.contour(Xc, Yc, pdf,levels=levels,colors='red',linewidths=2)
fmt = {
    levels[2]: r"1$\sigma$",
    levels[1]: r"2$\sigma$",
    levels[0]: r"3$\sigma$"
}

plt.clabel(contours, inline=True, fmt=fmt, fontsize=12)
# Styling
plt.title("2D Histogram + Contour")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--')


plt.show()
```
