```python
#4.3
import numpy as np
rng=np.random.default_rng()
w=input("input R/P/S for rock/paper/scissors",)
rand=int(rng.integers(low=0, high=3))

if w=="R":#Rock
    w=0
    if w==rand:
        print("RdR DRAW")
    elif (w+1)==rand:
        print("PbR LOSS")
    else:
        print("RbS WIN")
        
elif w=="P": #PAPER
    w=1
    if w==rand:
        print("PdP DRAW")
    elif (w+1)==rand:
        print("SbP LOSS")
    else:
        print("PbR WIN")
        
elif w=="S": #SCISSORS
    w=2
    if w==rand:
        print("SdS DRAW")
    elif (w-2)==rand:
        print("RbS LOSS")
    else:
        print("SbP WIN")
```

```python
#4.4
import numpy as np
import scipy
import matplotlib.pyplot as plt
mean=185
std=8
L=np.random.normal(mean,std,100000)
plt.hist(L,100)
plt.axvline(mean,0,1, color="r",label=f'mean={np.mean(L)}')
plt.axvspan(mean-std,mean+std,0,1,alpha=0.4, color="y", label=f'std={np.std(L)}')
plt.legend()
print(f'The mean is {np.mean(L)} and the std is {np.std(L)}')

h=input("Calculate the probability of height being below h, h=",)
heightinput=scipy.stats.norm.cdf(float(h),np.mean(L),np.std(L))
withinstd=scipy.stats.norm.cdf(np.mean(L)+np.std(L),np.mean(L),np.std(L))-scipy.stats.norm.cdf(np.mean(L)-np.std(L),np.mean(L),np.std(L))
print ()
print (heightinput)
print(withinstd)
```
